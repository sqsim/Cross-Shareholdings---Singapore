```python
def standardised_name(old_namelist):
    # PRIVATE
    pte_vars = ["(PTE)","(PRIVATE)","PTE"]
    # LIMITED
    ltd_vars = ["(LTD)","(LIMITED)","LTD"]
    # SINGAPORE
    SG_vars = ["(S)","(SPORE)","(S'PORE)","(SINGAPORE)","S'PORE","SPORE"]
    # NOMINEES
    Noms_vars = ["(NOMS)","NOMS"]
    # HOLDINGS
    Holdings_vars = ["(HLDGS)","(HOLDINGS)","HLDGS"]
#     COMPANY
    Company_vars = ["(CO)","CO"]
#   PRIVATE LIMITED 
    pl_vars = ["(PL)","PL"]

    name_list = old_namelist.copy()
    print("===========================")
    # print(name_list)
    for names in name_list:
        new_names = re.sub("[\.\,]","",names)
    #     print(new_names)
        index = name_list.index(names)
    #     print(index)
        name_list[index] = new_names

    # print(name_list)


#     print(type(name_list))
    for names in name_list:
        split_names = names.split()
        for pte in pte_vars:
            for splited in split_names:
                if splited == pte:
                    splited_index = split_names.index(splited)
                    split_names[splited_index] = "PRIVATE"
                    print(pte)
                    print(names)
                    x = " ".join(split_names)
                    index = name_list.index(names)
                    name_list[index] = x
                    print(x)
                    print("=============")
                    break

#     print(name_list)
    for names in name_list:
        split_names = names.split()
        for ltd in ltd_vars:
            for splited in split_names:
                if splited == ltd:
                    splited_index = split_names.index(splited)
                    split_names[splited_index] = "LIMITED"
                    print(ltd)
                    print(names)
                    x = " ".join(split_names)
                    index = name_list.index(names)
                    name_list[index] = x
                    print(x)
                    print("=============")
                    break
                    
#     print(name_list)
    for names in name_list:
        split_names = names.split()
        for nomms in SG_vars:
            for splited in split_names:
                if splited == nomms:
                    splited_index = split_names.index(splited)
                    split_names[splited_index] = "SINGAPORE"
                    print(nomms)
                    print(names)
                    x = " ".join(split_names)
                    index = name_list.index(names)
                    name_list[index] = x
                    print(x)
                    print("=============")
                    break
#     print(name_list)

    for names in name_list:
        split_names = names.split()
        for nomms in Noms_vars:
            for splited in split_names:
                if splited == nomms:
                    splited_index = split_names.index(splited)
                    split_names[splited_index] = "NOMINEES"
                    print(hold)
                    print(names)
                    x = " ".join(split_names)
                    index = name_list.index(names)
                    name_list[index] = x
                    print(x)
                    print("=============")
                    break
                
#     print(name_list)
    for names in name_list:
        split_names = names.split()
        for hold in Holdings_vars:
            for splited in split_names:
                if splited == hold:
                    splited_index = split_names.index(splited)
                    split_names[splited_index] = "HOLDINGS"
                    print(hold)
                    print(names)
                    x = " ".join(split_names)
                    index = name_list.index(names)
                    name_list[index] = x
                    print(x)
                    print("=============")
                    break
                
                
    for names in name_list:
        split_names = names.split()
        for com in Company_vars:
            for splited in split_names:
                if splited == com:
                    splited_index = split_names.index(splited)
                    split_names[splited_index] = "COMPANY"
                    print(com)
                    print(names)
                    x = " ".join(split_names)
                    index = name_list.index(names)
                    name_list[index] = x
                    print(x)
                    print("=============")
                    break
    print(name_list)
    for names in name_list:
        split_names = names.split()
        for pl in pl_vars:
            for splited in split_names:
                if splited == com:
                    splited_index = split_names.index(splited)
                    split_names[splited_index] = "PRIVATE LIMITED"
                    print(pl)
                    print(names)
                    x = " ".join(split_names)
                    index = name_list.index(names)
                    name_list[index] = x
                    print(x)
                    print("=============")
                    break
    print(name_list)
    return name_list
```
