### Remove Noise mostly in column or no noise
<div class="alert alert-block alert-danger">
can be used to loop through all the uncleaned csv or individually ,as all the files have different type of noise use with caution,the code needs to change according to the noise

```python
number = 28
range_l = number+1
length = len(company_list)
#  can be used to loop through all the uncleaned csv or individually base on range
for i in range(number ,range_l):
    number = i
    print("{} / {}".format(number, length))
    first_col = "NAME"
    second_col = "SHARES"
    third_col = "PERCENTAGE"

    name = company_list[number]
    path = directory+name+" shares.csv"
    print(path)
    df =pd.read_csv(path)
    
    shape =df.shape
#     df = df.iloc[8:]
#   dropna
    df.dropna(axis = 1,thresh  = 1, inplace=True)
    df.dropna(axis = 0,thresh = 1, inplace=True)
    
    drop_col = 0

#   if the have more than 20 row drop the rest from behind
    
#     df.fillna("0",inplace=True)
#     df[df.columns[0]] = df[df.columns[0]] + df[df.columns[1]]
    drop_amt = df.shape[0] - 20
    if drop_amt > 0:
        df.drop(df.tail(drop_amt).index,inplace=True)
    display(df)
    print("===")
    

#   if the number of column more than 3 drop first column
    number_column = len(df.columns)
    print(number_column)
    if number_column == 1:
        name_list = []
        shareholder_list = []
        percentage_list = []
        for column in df:
            if df[column].dtypes == "object":
                sum = df[column].str.contains("\n").sum()
                if sum == 20:
                    print(df[column].str.contains("\n").any())
                    for item in df[column]:
                        spilt_list = item.split("\n",2)
                        name_list.append(spilt_list[0])
                        shareholder_list.append(spilt_list[1])
                        percentage_list.append(spilt_list[2])
                    df.drop([column],axis = 1 ,inplace =True)
                    df["new col 1"] = name_list
                    df["new col 2"] = shareholder_list
                    df["new col 3"] = percentage_list
                    break
                    


#     shareholder_list = []
#     percentage_list = []
#     for column in df:
#         if df[column].dtypes == "object":
#             sum = df[column].str.contains("\n").sum()
#             if sum == 21:
#                 print(df[column].str.contains("\n").any())
#                 for item in df[column]:
#                     spilt_list = item.split("\n",1)
#                     shareholder_list.append(spilt_list[0])
#                     percentage_list.append(spilt_list[1])
#                 df.drop([column],axis = 1 ,inplace =True)
#                 df["new col 1"] = shareholder_list
#                 df["new col 2"] = percentage_list
#                 break

#     display(df)
#     df = df[[df.columns[1],df.columns[4],df.columns[5]]]
#     df.dropna(axis=0,inplace=True)

#     display(df)
    number_column = len(df.columns)
    while number_column > 3:
        df.drop(df.columns[drop_col], axis=1 ,inplace=True)
        number_column = len(df.columns)
        print(number_column)
    display(df) 
    

    if df[df.columns[0]].str.contains("[\d+]").any():
        print('contains number or/and newline')
        df[df.columns[0]] = df[df.columns[0]].str.replace('[\d+\n,]', '')
    
   
    print("............................")
#     display(df)
    
    if "CULM. %" in df.columns:
        df.drop("CULM. %" , axis= 1 , inplace = True)
#     rename columns
    df.columns = [first_col , second_col , third_col]
#     replace \t into space
    
    df[first_col] = df[first_col].str.replace("\t" , " ")
    df[second_col] = df[second_col].str.replace("[\n%]" , "")
#     replace \n as empty
    df[third_col] = df[third_col].astype(str)
    
    if df[third_col].str.contains("[\n%]").any():
        df[third_col] = df[third_col].str.replace("[\n%]" , "")
    for column in df:
        df[column] = df[column].str.strip()
    

    df.reset_index(inplace=True,drop=True)
    display(df)
    new_directory = "./cleaned_shareholder/"
    saveto = new_directory+name+" SHARES.csv"
    print(saveto)
    df.to_csv(saveto,index =False)
```

### Remove NANs in cleaned csv
There may still contain nans in the csv due to deleted columns and rows

```python
new_directory = "./cleaned_shareholder/"
for filename in os.listdir(new_directory):
    if filename.endswith(".csv"):
        path = new_directory+filename
        print(path)
        df = pd.read_csv(path,encoding = "ISO-8859–1")
        df.dropna(inplace=True)
        df.to_csv(path,index=False,encoding = "utf-8")
```
